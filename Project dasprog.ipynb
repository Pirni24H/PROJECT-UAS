```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ump.csv")
df
```

```python
plt.figure(figsize=(14,6))
plt.bar(df['provinsi'], df['ump.2022'], color='pink')
plt.title('Upah Miinimum provinsi di Tahun 2020')
plt.xlabel('Provinsi')
plt.ylabel('Upah Minimum(Rupiah)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12,8))

for tahun in ['ump.2006', 'ump.2007', 'ump.2008', 'ump.2009', 'ump.2010', 'ump.2011', 'ump.2012', 'ump.2013', 'ump.2014', 'ump.2015']:
    plt.plot(df['provinsi'], df[tahun], marker='o',label=tahun)

plt.title('Upah Minimum Provinsi Tahun 2006 - 2015')
plt.xlabel('Provinsi')
plt.ylabel("Upah Minimum(Rupiah)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.title('Presentase Upah Minimum Provinsi tahun 2022')
plt.pie(df['ump.2022'],labels=df['provinsi'], startangle=90, autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```
